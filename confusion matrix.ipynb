```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from tensorflow.keras.datasets import cifar10
from tensorflow.keras.models import load_model
from sklearn.metrics import confusion_matrix

(_, _), (x_test, y_test) = cifar10.load_data()

x_test = x_test.astype('float32') / 255.0

y_test = y_test.flatten()

model = load_model(r'C:\Users\Clay\Desktop\MN\Biometric\TP-CNN\all model\resnet20V3.h5')


predictions = model.predict(x_test)

predicted_classes = np.argmax(predictions, axis=1)

cm = confusion_matrix(y_test, predicted_classes)

classes = ['airplane', 'automobile', 'bird', 'cat', 'deer', 
           'dog', 'frog', 'horse', 'ship', 'truck']

plt.figure(figsize=(10, 8))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', 
            xticklabels=classes, yticklabels=classes)
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.title('Confusion Matrix for CIFAR-10 - Resnet20 V3 144Epoch')

plt.savefig('C:\\Users\\Clay\\Desktop\\MN\\Biometric\\TP-CNN\\confusion matrix\\resnet20V3.png', 
            dpi=300, bbox_inches='tight')

plt.show()
```
